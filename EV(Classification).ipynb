# filling datas in missing columns


```python
col=ev.columns
for col_name in col:
  if ev[col_name].dtypes=='object':
    ev[col_name]=ev[col_name].fillna(ev[col_name].mode()[0])
  else:
    ev[col_name]=ev[col_name].fillna(ev[col_name].mean())
```

##FINDING OUTLIERS


```python
# checking the outliers
plt.figure(figsize=(40,25),facecolor='white')
plotnumber=1

for column in ev:
  if plotnumber<=9:
    ax=plt.subplot(3,3,plotnumber)
    sns.boxplot(ev[column])
    plt.xlabel(column,fontsize=20)
    plt.ylabel('Count',fontsize=20)
    plotnumber+=1
  plt.tight_layout()
```

```python
ev.duplicated().sum()
```

##HANDLING THE OUTLIERS


```python
# Handling the outliers using IQR
numeric_cols = ev.select_dtypes(include='number')
for col in numeric_cols:
    Q1 = ev[col].quantile(0.25)
    Q3 = ev[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    ev = ev[(ev[col] >= lower) & (ev[col] <= upper)]
```

```python
X=ev.drop('Electric Vehicle Type',axis=1)
y=ev['Electric Vehicle Type']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.80,random_state=42)
```

```python
print("Value counts in original data:")
print(y.value_counts())

print("\nValue counts in training data:")
print(y_train.value_counts())
```

```python
y_test.value_counts()
```

```python
type(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train.astype(int))
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(y_test.astype(int),y_pred)
```

```python
y_pred
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import *
# create a decision tree classifier with class weights
class_weight = {0:1.0, 1:2.0} # assigning twice the weight to survived=1
model = DecisionTreeClassifier(class_weight=class_weight)

model.fit(X_train,y_train.astype(int))
y_pred=model.predict(X_test)
accuracy = accuracy_score(y_test.astype(int),y_pred)
print('Accuracy:',accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X_train,y_train.astype(int))
y_pred=model.predict(X_test)
accuracy = accuracy_score(y_test.astype(int),y_pred)
print('Accuracy:',accuracy)
```

```python
# Display Confusion Matrix for Logistic Regression
sns.heatmap(confusion_matrix(y_test.astype(int),y_pred),annot=True,fmt='g')
plt.title('Confusion Matrix - Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Display Classification Report for Logistic Regression
print(classification_report(y_test.astype(int),y_pred))
```
