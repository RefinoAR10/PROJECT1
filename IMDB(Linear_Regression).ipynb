```python
year_counts = imdb['Release'].value_counts().sort_index()
fig, ax = plt.subplots(figsize = (12,6))
bars = ax.bar(year_counts.index, year_counts.values)

ax.set_xlabel("Release Year", fontsize = 10)
ax.set_ylabel("Frequency", fontsize = 10)
ax.set_title("No. of Movies Released in a Year", fontsize = 14)
ax.set_xticks(year_counts.index)

ax.tick_params(axis = 'x', labelsize = 8, rotation = 90)
ax.tick_params(axis = 'y', labelsize = 10)

# data labels
for b in bars:
    h = b.get_height()
    ax.text(b.get_x() + b.get_width()/2, h, str(int(h)),
            ha = 'center', va = 'bottom', fontsize = 8)

    fig.tight_layout()
plt.show()
```

```python
# average rating for a specific release year

# grouping by 'Release' year and calculating average rating
avg_rating_per_year = imdb.groupby('Release')['Ratings'].mean()

# sorting by year (good for proper x-axis order)
avg_rating_per_year = avg_rating_per_year.sort_index()

plt.figure(figsize = (12,6))
plt.plot(avg_rating_per_year.index, avg_rating_per_year.values, marker = 'o', linestyle = '-')
plt.xticks(avg_rating_per_year.index, rotation = 90, fontsize = 8)
plt.yticks(fontsize = 10)
plt.xlabel("Release Year", fontsize = 10)
plt.ylabel("Average Rating", fontsize = 10)
plt.title("Average Rating for a Specific Release Year", fontsize = 14)

# data labels
for x, y in zip(avg_rating_per_year.index, avg_rating_per_year.values):
    plt.text(x, y+0.01, f"{y:.2f}", ha = 'center', va = 'bottom', fontsize = 8)

plt.tight_layout()
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

x=imdb[imdb.columns[::-1]]
y=imdb['Ratings']

vif_df=pd.DataFrame({
    'features':x.columns,
    'vif':[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
})
print(vif_df.sort_values('vif',ascending=False))
```

```python
# model creation
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.75,random_state=42)
x_train
```

```python
# to create or building the model
l_model=LinearRegression()
l_model.fit(x_train,y_train)
```

```python

y_pred=l_model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_on_test=r2_score(y_test,y_pred)
r2_on_test
```

```python
weights=l_model.coef_
intercept=l_model.intercept_
print(weights)
print(intercept)
x_test
```
